```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)



df = pd.read_csv(r"C:\Users\User\Downloads\archive (3)\movies.csv")
```

```python
df.head()
```

```python
#checking for missing data 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
#checking for datatypes for columns
df.dtypes
```

```python
# change Nan to 0
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')



```

```python
df_Numeric = df
for col_name in df_Numeric.columns:
    if(df_Numeric[col_name].dtype == 'object'):
        df_Numeric[col_name]=df_Numeric[col_name].astype('category')
        df_Numeric[col_name]=df_Numeric[col_name].cat.codes

df_Numeric
    
```

```python
df
```

```python
correlation_matrix =  df_Numeric.corr(numeric_only=True,method='pearson') 
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Budget For Film')
plt.show()
```
