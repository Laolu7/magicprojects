```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv(r"C:\Users\User\Downloads\archive (4)\CountryHits240901.csv")
df
```

```python
# checking information to understand the structure
df_head = df.head()
df_info = df.info()
df_head, df_info
```

```python
#  data cleaning
# Convert 'Release' column to datetime format
df['Release'] = pd.to_datetime(df['Release'], errors='coerce')

```

```python
# Checking for missing values to decide on handling them
missing_values = df.isnull().sum()

```

```python
# Drop duplicates if any
df_cleaned = df.drop_duplicates()

# checking for duplicate removal and inspecting missing data counts
cleaned_summary = df_cleaned.info()
missing_values, cleaned_summary

```

```python
# Exploring the data

# Statistical summary of numerical columns
df_summary = df_cleaned.describe()

# Checking unique values in categorical columns
unique_artists = df_cleaned['Artist1'].nunique()
unique_albums = df_cleaned['Album'].nunique()

# Correlation matrix for numerical features
correlation_matrix = df_cleaned.corr(numeric_only=True)

df_summary, unique_artists, unique_albums, correlation_matrix





```

```python

sns.set_theme(style="whitegrid")

# Visualization: Top 10 artists with the most entries
top_artists = df_cleaned['Artist1'].value_counts().head(10)

plt.figure(figsize=(10, 6))
# Assign y to hue and set legend=False
sns.barplot(x=top_artists.values, y=top_artists.index, palette='viridis', hue=top_artists.index, legend=False)
plt.title("Top 10 Artists by Song Count")
plt.xlabel("Number of Songs")
plt.ylabel("Artist")
plt.show()

```

```python
#  Visualization :Top 10 albums with the most entries
top_albums = df_cleaned['Album'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_albums.values, y=top_albums.index, palette='rocket', hue=top_artists.index, legend=False)
plt.title("Top 10 Albums by Song Count")
plt.xlabel("Number of Songs")
plt.ylabel("Album")
plt.show()
```

```python
# Visualization : average popularity by artist and sort to get top 10
avg_popularity_by_artist = df_cleaned.groupby('Artist1')['Popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_popularity_by_artist.values, y=avg_popularity_by_artist.index, palette='mako', hue=top_artists.index, legend=False)
plt.title("Top 10 Artists by Average Popularity")
plt.xlabel("Average Popularity")
plt.ylabel("Artist")
plt.show()
```

```python
#Visualization : Distrubution of song duration
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['Duration'], bins=20, kde=True, color='purple')
plt.title("Distribution of Song Duration")
plt.xlabel("Duration (ms)")
plt.ylabel("Frequency")
plt.show()

```

```python
#Visualization : popularity vs danceability
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='Danceability', y='Popularity', hue='Energy', palette='cool', size='Energy', sizes=(20, 200))
plt.title("Popularity vs Danceability")
plt.xlabel("Danceability")
plt.ylabel("Popularity")
plt.legend(title="Energy Level", loc='upper left')
plt.show()
```
